# Histogram

```python
import cv2
import matplotlib.pyplot as plt
```

## Mengambil Gambar dan Mengubahnya ke Grayscale

```python
img = cv2.imread("kopiGrayscale.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGRA2GRAY)
plt.imshow(gray)
plt.show()
```

## Menampilkan Histogram

```python
plt.hist(gray.ravel(), 256, [0, 256])
plt.show()
```
